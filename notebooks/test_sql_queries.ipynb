```python
from sqlalchemy import create_engine 
```

```python
# test de requete SQL
#Create engine to connect to the newIMDB database
engineIMDB = create_engine('sqlite:///../data/db/newIMDB.db')
connIMDB = engineIMDB.connect()

result = connIMDB.execute("SELECT title_id,rating,votes" 
                          " FROM ratings"  
                          " WHERE votes > 5000" 
                          " ORDER BY rating" 
                          " DESC, ratings.votes" 
                          " DESC LIMIT 5")
result.fetchall()
```

```python
result2 = connIMDB.execute("SELECT primary_title, rating, votes" 
                          " FROM titles" 
                          " INNER JOIN ratings" 
                          " ON titles.title_id=ratings.title_id" 
                          " WHERE votes > 5000" 
                          " ORDER BY ratings.rating" 
                          " DESC, ratings.votes" 
                          " DESC LIMIT 10")
result2.fetchall()
```
